```
def model():  
  vision_lambda = Input(shape=(12,128,256))
  vision_per = Permute((2,3,1))(vision_lambda)
  vision_conv2d = Conv2D(16,(5,5),strides=(1,1), padding ='same',input_shape = vision_per.shape)(vision_per)
  vision_elu = ELU()(vision_conv2d)
  vision_maxpool2d = AveragePooling2D(pool_size=(3, 3), strides=(2,2), padding='same')(vision_elu)
  vision1_con2d_l_1 = Conv2D(64,(1,1),strides=(1,1), padding='valid')(vision_maxpool2d)
  vision1_con2d_l_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision1_con2d_l_1)
  vision1_con2d_r_1 = Conv2D(16,(1,1),strides=(1,1), padding='valid')(vision_maxpool2d)
  vision1_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision1_con2d_r_1)
  vision1_con2d_r_2 = Activation('elu')(vision1_con2d_r_1_batch)
  vision1_con2d_r_3 = Conv2D(16,(3,3),strides=(1,1), padding='same')(vision1_con2d_r_2)
  vision1_con2d_r_3_batch  = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision1_con2d_r_3)
  vision1_con2d_r_4 = Activation('elu')(vision1_con2d_r_3_batch)
  vision1_con2d_r_4 = Conv2D(64,(1,1),strides=(1,1), padding='valid')(vision1_con2d_r_4)
  vision1_con2d_r_4_batch  = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision1_con2d_r_4)
  vision1_con2d_add = Add()([vision1_con2d_r_4_batch,vision1_con2d_l_1_batch])
  vision1_con2d_elu = Activation('elu')(vision1_con2d_add)

  vision2_con2d_r_1 = Conv2D(16,(1,1),strides=(1,1), padding='valid')(vision1_con2d_elu)
  vision2_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision2_con2d_r_1)
  vision2_con2d_r_1_active = Activation('elu')(vision2_con2d_r_1_batch)
  vision2_con2d_r_2 = Conv2D(16,(3,3),strides=(1,1), padding='same')(vision2_con2d_r_1_active)
  vision2_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision2_con2d_r_2)
  vision2_con2d_r_2_active  = Activation('elu')(vision2_con2d_r_2_batch)
  vision2_con2d_r_3 = Conv2D(64,(1,1),strides=(1,1), padding='valid')(vision2_con2d_r_2_active)
  vision2_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision2_con2d_r_3)

  vision2_con2d_add = Add()([vision1_con2d_elu,vision2_con2d_r_3_batch])
  vision2_con2d_elu = Activation('elu')(vision2_con2d_add)

  vision3_con2d_r_1 = Conv2D(16,(1,1),strides=(1,1), padding='valid')(vision2_con2d_elu)
  vision3_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision3_con2d_r_1)
  vision3_con2d_r_1_active = Activation('elu')(vision3_con2d_r_1_batch)
  vision3_con2d_r_2 = Conv2D(16,(3,3),strides=(1,1), padding='same')(vision3_con2d_r_1_active)
  vision3_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision3_con2d_r_2)
  vision3_con2d_r_2_active  = Activation('elu')(vision3_con2d_r_2_batch)
  vision3_con2d_r_3 = Conv2D(64,(1,1),strides=(1,1), padding='valid')(vision3_con2d_r_2_active)
  vision3_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision3_con2d_r_3)

  vision3_con2d_add = Add()([vision2_con2d_elu,vision3_con2d_r_3_batch])
  vision3_con2d_elu = Activation('elu')(vision3_con2d_add)

  vision4_con2d_r_1 = Conv2D(32,(1,1),strides=(2,2), padding='valid')(vision3_con2d_elu)
  vision4_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision4_con2d_r_1)
  vision4_con2d_r_1_active = Activation('elu')(vision4_con2d_r_1_batch)
  vision4_con2d_r_2 = Conv2D(32,(3,3),strides=(1,1), padding='same')(vision4_con2d_r_1_active)
  vision4_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision4_con2d_r_2)
  vision4_con2d_r_2_active  = Activation('elu')(vision4_con2d_r_2_batch)
  vision4_con2d_r_3 = Conv2D(128,(1,1),strides=(1,1), padding='valid')(vision4_con2d_r_2_active)
  vision4_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision4_con2d_r_3)
  vision4_con2d_l_1 = Conv2D(128,(1,1),strides=(2,2), padding='valid')(vision3_con2d_elu)
  vision4_con2d_l_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision4_con2d_l_1)

  vision4_con2d_add = Add()([vision4_con2d_l_1_batch,vision4_con2d_r_3_batch])
  vision4_con2d_elu = Activation('elu')(vision4_con2d_add)

  vision5_con2d_r_1 = Conv2D(32,(1,1),strides=(1,1), padding='valid')(vision4_con2d_elu)
  vision5_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision5_con2d_r_1)
  vision5_con2d_r_1_active = Activation('elu')(vision5_con2d_r_1_batch)
  vision5_con2d_r_2 = Conv2D(32,(3,3),strides=(1,1), padding='same')(vision5_con2d_r_1_active)
  vision5_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision5_con2d_r_2)
  vision5_con2d_r_2_active  = Activation('elu')(vision5_con2d_r_2_batch)
  vision5_con2d_r_3 = Conv2D(128,(1,1),strides=(1,1), padding='valid')(vision5_con2d_r_2_active)
  vision5_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision5_con2d_r_3)

  vision5_con2d_add = Add()([vision4_con2d_elu,vision5_con2d_r_3_batch])
  vision5_con2d_elu = Activation('elu')(vision5_con2d_add)

  vision6_con2d_r_1 = Conv2D(32,(1,1),strides=(1,1), padding='valid')(vision5_con2d_elu)
  vision6_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision6_con2d_r_1)
  vision6_con2d_r_1_active = Activation('elu')(vision6_con2d_r_1_batch)
  vision6_con2d_r_2 = Conv2D(32,(3,3),strides=(1,1), padding='same')(vision6_con2d_r_1_active)
  vision6_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision6_con2d_r_2)
  vision6_con2d_r_2_active  = Activation('elu')(vision6_con2d_r_2_batch)
  vision6_con2d_r_3 = Conv2D(128,(1,1),strides=(1,1), padding='valid')(vision6_con2d_r_2_active)
  vision6_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision6_con2d_r_3)

  vision6_con2d_add = Add()([vision5_con2d_elu,vision6_con2d_r_3_batch])
  vision6_con2d_elu = Activation('elu')(vision6_con2d_add)

  vision7_con2d_r_1 = Conv2D(32,(1,1),strides=(1,1), padding='valid')(vision6_con2d_elu)
  vision7_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision7_con2d_r_1)
  vision7_con2d_r_1_active = Activation('elu')(vision7_con2d_r_1_batch)
  vision7_con2d_r_2 = Conv2D(32,(3,3),strides=(1,1), padding='same')(vision7_con2d_r_1_active)
  vision7_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision7_con2d_r_2)
  vision7_con2d_r_2_active  = Activation('elu')(vision7_con2d_r_2_batch)
  vision7_con2d_r_3 = Conv2D(128,(1,1),strides=(1,1), padding='valid')(vision7_con2d_r_2_active)
  vision7_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision7_con2d_r_3)

  vision7_con2d_add = Add()([vision6_con2d_elu,vision7_con2d_r_3_batch])
  vision7_con2d_elu = Activation('elu')(vision6_con2d_add)

  vision8_con2d_r_1 = Conv2D(48,(1,1),strides=(2,2), padding='valid')(vision7_con2d_elu)
  vision8_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision8_con2d_r_1)
  vision8_con2d_r_1_active = Activation('elu')(vision8_con2d_r_1_batch)
  vision8_con2d_r_2 = Conv2D(48,(3,3),strides=(1,1), padding='same')(vision8_con2d_r_1_active)
  vision8_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision8_con2d_r_2)
  vision8_con2d_r_2_active  = Activation('elu')(vision8_con2d_r_2_batch)
  vision8_con2d_r_3 = Conv2D(192,(1,1),strides=(1,1), padding='valid')(vision8_con2d_r_2_active)
  vision8_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision8_con2d_r_3)
  vision8_con2d_l_1 = Conv2D(192,(1,1),strides=(2,2), padding='valid')(vision7_con2d_elu)
  vision8_con2d_l_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision8_con2d_l_1)

  vision8_con2d_add = Add()([vision8_con2d_l_1_batch,vision8_con2d_r_3_batch])
  vision8_con2d_elu = Activation('elu')(vision8_con2d_add)

  vision9_con2d_r_1 = Conv2D(48,(1,1),strides=(1,1), padding='valid')(vision8_con2d_elu)
  vision9_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision9_con2d_r_1)
  vision9_con2d_r_1_active = Activation('elu')(vision9_con2d_r_1_batch)
  vision9_con2d_r_2 = Conv2D(48,(3,3),strides=(1,1), padding='same')(vision9_con2d_r_1_active)
  vision9_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision9_con2d_r_2)
  vision9_con2d_r_2_active  = Activation('elu')(vision9_con2d_r_2_batch)
  vision9_con2d_r_3 = Conv2D(192,(1,1),strides=(1,1), padding='valid')(vision9_con2d_r_2_active)
  vision9_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision9_con2d_r_3)

  vision9_con2d_add = Add()([vision8_con2d_elu,vision9_con2d_r_3_batch])
  vision9_con2d_elu = Activation('elu')(vision9_con2d_add)

  vision10_con2d_r_1 = Conv2D(48,(1,1),strides=(1,1), padding='valid')(vision9_con2d_elu)
  vision10_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision10_con2d_r_1)
  vision10_con2d_r_1_active = Activation('elu')(vision10_con2d_r_1_batch)
  vision10_con2d_r_2 = Conv2D(48,(3,3),strides=(1,1), padding='same')(vision10_con2d_r_1_active)
  vision10_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision10_con2d_r_2)
  vision10_con2d_r_2_active  = Activation('elu')(vision10_con2d_r_2_batch)
  vision10_con2d_r_3 = Conv2D(192,(1,1),strides=(1,1), padding='valid')(vision10_con2d_r_2_active)
  vision10_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision10_con2d_r_3)

  vision10_con2d_add = Add()([vision9_con2d_elu,vision10_con2d_r_3_batch])
  vision10_con2d_elu = Activation('elu')(vision10_con2d_add)


  vision11_con2d_r_1 = Conv2D(48,(1,1),strides=(1,1), padding='valid')(vision10_con2d_elu)
  vision11_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision11_con2d_r_1)
  vision11_con2d_r_1_active = Activation('elu')(vision11_con2d_r_1_batch)
  vision11_con2d_r_2 = Conv2D(48,(3,3),strides=(1,1), padding='same')(vision11_con2d_r_1_active)
  vision11_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision11_con2d_r_2)
  vision11_con2d_r_2_active  = Activation('elu')(vision11_con2d_r_2_batch)
  vision11_con2d_r_3 = Conv2D(192,(1,1),strides=(1,1), padding='valid')(vision11_con2d_r_2_active)
  vision11_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision11_con2d_r_3)

  vision11_con2d_add = Add()([vision10_con2d_elu,vision11_con2d_r_3_batch])
  vision11_con2d_elu = Activation('elu')(vision11_con2d_add)

  vision12_con2d_r_1 = Conv2D(48,(1,1),strides=(1,1), padding='valid')(vision11_con2d_elu)
  vision12_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision12_con2d_r_1)
  vision12_con2d_r_1_active = Activation('elu')(vision12_con2d_r_1_batch)
  vision12_con2d_r_2 = Conv2D(48,(3,3),strides=(1,1), padding='same')(vision12_con2d_r_1_active)
  vision12_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision12_con2d_r_2)
  vision12_con2d_r_2_active  = Activation('elu')(vision12_con2d_r_2_batch)
  vision12_con2d_r_3 = Conv2D(192,(1,1),strides=(1,1), padding='valid')(vision12_con2d_r_2_active)
  vision12_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision12_con2d_r_3)

  vision12_con2d_add = Add()([vision11_con2d_elu,vision12_con2d_r_3_batch])
  vision12_con2d_elu = Activation('elu')(vision12_con2d_add)

  vision13_con2d_r_1 = Conv2D(48,(1,1),strides=(1,1), padding='valid')(vision12_con2d_elu)
  vision13_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision13_con2d_r_1)
  vision13_con2d_r_1_active = Activation('elu')(vision13_con2d_r_1_batch)
  vision13_con2d_r_2 = Conv2D(48,(3,3),strides=(1,1), padding='same')(vision13_con2d_r_1_active)
  vision13_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision13_con2d_r_2)
  vision13_con2d_r_2_active  = Activation('elu')(vision13_con2d_r_2_batch)
  vision13_con2d_r_3 = Conv2D(192,(1,1),strides=(1,1), padding='valid')(vision13_con2d_r_2_active)
  vision13_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision13_con2d_r_3)

  vision13_con2d_add = Add()([vision12_con2d_elu,vision13_con2d_r_3_batch])
  vision13_con2d_elu = Activation('elu')(vision13_con2d_add)


  vision14_con2d_r_1 = Conv2D(64,(1,1),strides=(2,2), padding='valid')(vision13_con2d_elu)
  vision14_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision14_con2d_r_1)
  vision14_con2d_r_1_active = Activation('elu')(vision14_con2d_r_1_batch)
  vision14_con2d_r_2 = Conv2D(64,(3,3),strides=(1,1), padding='same')(vision14_con2d_r_1_active)
  vision14_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision14_con2d_r_2)
  vision14_con2d_r_2_active  = Activation('elu')(vision14_con2d_r_2_batch)
  vision14_con2d_r_3 = Conv2D(256,(1,1),strides=(1,1), padding='valid')(vision14_con2d_r_2_active)
  vision14_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision14_con2d_r_3)
  vision14_con2d_l_1 = Conv2D(256,(1,1),strides=(2,2), padding='valid')(vision13_con2d_elu)
  vision14_con2d_l_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision14_con2d_l_1)

  vision14_con2d_add = Add()([vision14_con2d_l_1_batch,vision14_con2d_r_3_batch])
  vision14_con2d_elu = Activation('elu')(vision14_con2d_add)

  vision15_con2d_r_1 = Conv2D(64,(1,1),strides=(1,1), padding='valid')(vision14_con2d_elu)
  vision15_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision15_con2d_r_1)
  vision15_con2d_r_1_active = Activation('elu')(vision15_con2d_r_1_batch)
  vision15_con2d_r_2 = Conv2D(64,(3,3),strides=(1,1), padding='same')(vision15_con2d_r_1_active)
  vision15_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision15_con2d_r_2)
  vision15_con2d_r_2_active  = Activation('elu')(vision15_con2d_r_2_batch)
  vision15_con2d_r_3 = Conv2D(256,(1,1),strides=(1,1), padding='valid')(vision15_con2d_r_2_active)
  vision15_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision15_con2d_r_3)

  vision15_con2d_add = Add()([vision14_con2d_elu,vision15_con2d_r_3_batch])
  vision15_con2d_elu = Activation('elu')(vision15_con2d_add)

  vision16_con2d_r_1 = Conv2D(64,(1,1),strides=(1,1), padding='valid')(vision15_con2d_elu)
  vision16_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision16_con2d_r_1)
  vision16_con2d_r_1_active = Activation('elu')(vision16_con2d_r_1_batch)
  vision16_con2d_r_2 = Conv2D(64,(3,3),strides=(1,1), padding='same')(vision16_con2d_r_1_active)
  vision16_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision16_con2d_r_2)
  vision16_con2d_r_2_active  = Activation('elu')(vision16_con2d_r_2_batch)
  vision16_con2d_r_3 = Conv2D(256,(1,1),strides=(1,1), padding='valid')(vision16_con2d_r_2_active)
  vision16_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.00001001,center=True,scale=True)(vision16_con2d_r_3)

  vision16_con2d_add = Add()([vision15_con2d_elu,vision16_con2d_r_3_batch])
  vision16_con2d_elu = Activation('elu')(vision16_con2d_add)

  vision17_con2d_r_1 = Conv2D(96,(3,3),strides=(1,1), padding='same')(vision16_con2d_elu)
  vision17_con2d_r_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.001,center=True,scale=True)(vision17_con2d_r_1)
  vision17_con2d_r_1_active = ELU()(vision17_con2d_r_1_batch)
  vision17_con2d_r_2 = Conv2D(96,(3,3),strides=(1,1), padding='same')(vision17_con2d_r_1_active)
  vision17_con2d_r_2_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.001,center=True,scale=True)(vision17_con2d_r_2)
  vision17_con2d_r_2_active = ELU()(vision17_con2d_r_2_batch)
  vision17_con2d_r_2_Max = MaxPooling2D(pool_size=(2, 2), strides=(2,2), padding='valid', data_format= 'channels_last')(vision17_con2d_r_2_active)
  vision17_con2d_r_3 = Conv2D(128,(3,3),strides=(1,1), padding='same')(vision17_con2d_r_2_Max)
  vision17_con2d_r_3_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.001,center=True,scale=True)(vision17_con2d_r_3)
  vision17_con2d_r_3_active = ELU()(vision17_con2d_r_3_batch)
  vision17_con2d_r_4 = Conv2D(128,(3,3),strides=(1,1), padding='same')(vision17_con2d_r_3_active)
  vision17_con2d_r_4_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.001,center=True,scale=True)(vision17_con2d_r_4)
  vision17_con2d_r_4_active = ELU()(vision17_con2d_r_4_batch)
  vision17_con2d_r_4_Max = MaxPooling2D(pool_size=(2, 2), strides=(2,2), padding='valid', data_format= 'channels_last')(vision17_con2d_r_4_active)
  vision17_con2d_r_5 = Conv2D(192,(3,3),strides=(1,1), padding='same')(vision17_con2d_r_4_Max)
  vision17_con2d_r_5_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.001,center=True,scale=True)(vision17_con2d_r_5)
  vision17_con2d_r_5_active = ELU()(vision17_con2d_r_5_batch)
  vision17_con2d_r_6 = Conv2D(192,(3,3),strides=(1,1), padding='same')(vision17_con2d_r_5_active)
  vision17_con2d_r_6_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.001,center=True,scale=True)(vision17_con2d_r_6)
  vision17_con2d_r_6_active = ELU()(vision17_con2d_r_6_batch)
  flatten_r = Flatten()(vision17_con2d_r_6_active)
  dense_r = Dense(256)(flatten_r)
  vision17_batch = BatchNormalization(axis=1,momentum=0.99, epsilon=0.001,center=True,scale=True)(dense_r)
  vision17_Elu = ELU()(vision17_batch)
  vision17_con2d_l_1 = Conv2D(192,(3,3),strides=(1,1), padding='same')(vision16_con2d_elu)
  vision17_con2d_l_1_batch = BatchNormalization(axis=3,momentum=0.99, epsilon=0.001,center=True,scale=True)(vision17_con2d_l_1)
  vision17_con2d_l_1_active = ELU()(vision17_con2d_l_1_batch)
  flatten_l = Flatten()(vision17_con2d_l_1_active)
  concat_1 = Concatenate(axis= -1)([flatten_l,vision17_Elu])
  # first block concatenate on the right hand
  #r=right_hand
  #l=left_hand
  #***********right********************
  dense_con_r_1 = Dense(256)(concat_1)
  act_con_r_1 = Activation('relu')(dense_con_r_1)
  dense_con_r_2 = Dense(128)(act_con_r_1)
  act_con_r_2 = Activation('relu')(dense_con_r_2)
  dense_con_r_3 = Dense(64)(act_con_r_2)
  act_con_r_3 = Activation('relu')(dense_con_r_3)
  dense_con_r_4 = Dense(64)(act_con_r_3)
  act_con_r_4 = Activation('relu')(dense_con_r_4)
  dense_relu_r = Dense(4,activation='sigmoid')(act_con_r_4) 
  meta=dense_relu_r
  #***********left********************
  dense_relu_l = Dense(32)(act_con_r_4)
  dense_relu_l_reshape = Reshape((4,8))(dense_relu_l)
  dense_relu_l_soft = Softmax(axis=-1)(dense_relu_l_reshape)
  dense_relu_l_flatten = Flatten()(dense_relu_l_soft)
  dense_relu_l_act = Activation('relu')(dense_relu_l_flatten)
  dense_relu_l_dense = Dense(32)(dense_relu_l_act)
  snpe_pleaser2 = dense_relu_l_dense
  dense_con_l_1 = Dense(256)(concat_1)
  act_con_l_1 = ELU()(dense_con_l_1)
  dense_con_l_2 = Dense(128)(act_con_l_1)
  act_con_l_2 = ELU()(dense_con_l_2)
  dense_con_l_3 = Dense(64)(act_con_l_2)
  act_con_l_3 = Activation('linear')(dense_con_l_3)
  pose = act_con_l_3
  # second block concatenate on the left hand
  desire = Input(shape=(8,))
  dense_d = Dense(8)(desire)
  concat_2 = Concatenate(axis= -1)([concat_1,dense_d])
  dense_con = Dense(512)(concat_2)
  #right to left for dense 
  rnn_z = Dense(512)(dense_con)
  rnn_h = Dense(512)(dense_con)
  rnn_r = Dense(512)(dense_con)
  #right to left for rnn
  rnn_state = Input(shape=(1,512))
  rnn_state1 = Lambda(lambda x: squeeze(x, axis=0))(rnn_state)
  rnn_state2 = Lambda(lambda x: expand_dims(x, axis=1))(rnn_state1)
  rnn_rz = LSTM(512)(rnn_state2)
  snpe_pleaser = LSTM(512)(rnn_state2)
  rnn_rh = Dense(512)(snpe_pleaser)
  rnn_rr = LSTM(512)(rnn_state2)
  #right to left for operation
  add =  Add()([rnn_z,rnn_rz])
  act_add_1 = Activation('sigmoid')(add)
  act_add_1_d = Dense(512)(act_add_1)
  add_1 =  Add()([rnn_r,rnn_rr])
  act_add_2 = Activation('sigmoid')(add_1)
  multiply = Multiply()([act_add_2,rnn_rh])
  add_2 =  Add()([rnn_h,multiply])
  add_3_act = Activation('tanh')(add_2)
  multiply_1 = Multiply()([snpe_pleaser,act_add_1])
  multiply_2 = Multiply()([act_add_1_d,add_3_act])
  add_3 =  Add()([multiply_1,multiply_2])
  ##################path 
  dense_1_path = Dense(256)(add_3)  
  relu_1_path = ReLU()(dense_1_path)
  dense_2_path = Dense(192)(relu_1_path) 
  relu_2_path = ReLU()(dense_2_path)
  dense_3_path = Dense(128)(relu_2_path) 
  relu_3_path = ReLU()(dense_3_path)
  path = Dense(384)(relu_3_path) 
  #################left_lane
  dense_1_left = Dense(256)(add_3) 
  relu_1_left = ReLU()(dense_1_left)
  dense_2_left = Dense(192)(relu_1_left) 
  relu_2_left = ReLU()(dense_2_left)
  dense_3_left = Dense(128)(relu_2_left) 
  relu_3_left = ReLU()(dense_3_left)
  left_lane = Dense(385)(relu_3_left) 
  #################right_lane
  dense_1_right = Dense(256)(add_3) 
  relu_1_right = ReLU()(dense_1_right)
  dense_2_right = Dense(192)(relu_1_right) 
  relu_2_right = ReLU()(dense_2_right)
  dense_3_right = Dense(128)(relu_2_right) 
  relu_3_right = ReLU()(dense_3_right)
  right_lane = Dense(385)(relu_3_right) 
  ##################lead 
  dense_1_lead = Dense(256)(add_3)  
  relu_1_lead = ReLU()(dense_1_lead)
  dense_2_lead = Dense(192)(relu_1_lead) 
  relu_2_lead = ReLU()(dense_2_lead)
  dense_3_lead = Dense(128)(relu_2_lead) 
  relu_3_lead = ReLU()(dense_3_lead)
  lead = Dense(384)(relu_3_lead) 
  ################## long_x
  dense_1_long_x = Dense(256)(add_3) 
  relu_1_long_x = ReLU()(dense_1_long_x)
  dense_2_long_x = Dense(192)(relu_1_long_x) 
  relu_2_long_x = ReLU()(dense_2_long_x)
  dense_3_long_x = Dense(128)(relu_2_long_x) 
  relu_3_long_x = ReLU()(dense_3_long_x)
  long_x = Dense(58)(relu_3_long_x) 
  ################## long_v
  dense_1_long_v = Dense(256)(add_3) 
  relu_1_long_v = ReLU()(dense_1_long_v)
  dense_2_long_v = Dense(192)(relu_1_long_v) 
  relu_2_long_v = ReLU()(dense_2_long_v)
  dense_3_long_v = Dense(128)(relu_2_long_v) 
  relu_3_long_v = ReLU()(dense_3_long_v)
  long_v = Dense(200)(relu_3_long_v) 
  ################## long_a
  dense_1_long_a = Dense(256)(add_3) 
  relu_1_long_a = ReLU()(dense_1_long_a)
  dense_2_long_a = Dense(192)(relu_1_long_a) 
  relu_2_long_a = ReLU()(dense_2_long_a)
  dense_3_long_a = Dense(128)(relu_2_long_a) 
  relu_3_long_a = ReLU()(dense_3_long_a)
  long_a = Dense(200)(relu_3_long_a)
 
  model = Model(inputs=[vision_lambda,desire,rnn_state], outputs=[path,left_lane,right_lane,lead,long_x,long_v,long_a,meta,snpe_pleaser2,pose,add_3] )
  print(len(model.layers))

  model.compile(loss='categorical_crossentropy', optimizer='adam',  metrics=['accuracy'])
  return model,path,left_lane,right_lane,lead,long_x,long_v,long_a,meta,snpe_pleaser2,pose,add_3
model,path,left_lane,right_lane,lead,long_x,long_v,long_a,meta,snpe_pleaser2,pose,add_3 = model()  
```

```
def PreparedData(file):
  img_arrf= []
  ysf = []
  json_gt = [json.loads(line) for line in open(file,'r')]
  for i in range(0,300):
    gt = json_gt[i]              #### json_gt = each section from json file
    ys = gt['lanes']             ## label ys for fit model
    ysf.append(ys)      
    y_samples = gt['h_samples']
    xs = gt['raw_file']          ## input xs for fit model
    img = load_img('/content/drive/My Drive/'+xs)   ## input image
    img_arr = img_to_array(img)
    img_arr1 = img_arr.reshape((1,) + img_arr.shape)
    img_arr2 = img_arr1.reshape((1,) + img_arr1.shape)
    img_arrf.append(img_arr2)
  # img_f = img_to_array(img_a)

  for i in range(0,300):      ## convert label list into array
    ysf[i] = np.asarray(ysf[i])
    ysf[i] = np.resize(ysf[i],(1,64))
  for i in range(0,300):      ## resize each image to size of input model
    img_arrf[i] = np.resize(img_arrf[i],(1,12,128,256))
    
  return img_arrf,ysf         ## xs , ys array
```
